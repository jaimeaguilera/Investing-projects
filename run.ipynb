#### Study inclusion of alternative asset in the portfolio

```python
def normalize(weight_series,  w, alt_name):

    a = weight_series
    a = a*(1-w)
    a[alt_name] = w
    
    return a
```

```python
ifsw = Asset(ticker="IFSW.MI", name="MSCI World Multifactor", source="Yahoo")
alt_price = ifsw.get_asset()
alt_name = "IFSW"

alt_rets = alt_price.pct_change()["2016-09-28":]
r = pd.concat([portfolio.r.drop(columns = "PORT"), alt_rets], axis = 1)
r[alt_name] = r[alt_name].fillna(0)
col_drop_plot = ["PRUD", "BTC-EUR", "EUA", "ZPRS", "WCOA", "EUNZ", "4GLD", "FSMI", "BRYN"]
stats = portfolio.portfolio_report(r, date_beg = "2018", col_drop_plot = col_drop_plot)

roll_vol = portfolio.rolling_vol(r["2018":], col_drop_plot = col_drop_plot, time_period = 90)

weight_series = portfolio.w.iloc[-1]
weights = np.array([normalize(weight_series,  w, alt_name) for w in np.linspace(0,0.2,21)])

results = (r @ weights.T)
results.columns = np.around(weights[:,-1],2).tolist()
summary_stats(results["2018":], riskfree_rate = 0)
portfolio.rolling_corr(r["2018":], time_period = 90, col = 'EUNL')
```
